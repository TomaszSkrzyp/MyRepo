```python
import turtle as t
t.speed(10)
t.left(90)
```

```python
def rzulf(angle,side,minside):
    if side<minside:
        return 
    else:
        t.forward(side)
        t.rt(angle)
        rzulf(angle,side/2,minside)
        print(side)
        t.lt(2*angle)
        rzulf(angle,side/2,minside)
        t.rt(angle)
        t.forward(-side)
```

```python
rzulf(90,200,26)
```
