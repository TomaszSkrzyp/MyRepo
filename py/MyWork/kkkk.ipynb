```python
import numpy as np 
import pandas as pd
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_dataset=pd.read_csv("train.csv")
train_dataset.head()
```

```python
train_dataset.CryoSleep=train_dataset.CryoSleep.replace(True,1).replace(False,0)
train_dataset.VIP=train_dataset.VIP.replace(True,1).replace(False,0)
train_dataset.HomePlanet=train_dataset.HomePlanet.replace("Europa",1).replace("Mars",0).replace("Earth",2)

IDs=train_dataset.PassengerId
y=train_dataset.Transported
y=y.replace(True,1).replace(False,0)
train_dataset
```

```python
train_dataset=train_dataset.drop("Cabin", axis=1)
train_dataset=train_dataset.drop("Name", axis=1)
train_dataset=train_dataset.drop("Transported", axis=1)
train_dataset=train_dataset.drop("Destination", axis=1)
```

```python
train_dataset
```

```python
train_dataset.CryoSleep.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(train_dataset.iloc[:,1:])
train_dataset.iloc[:,1:]=imputer.transform(train_dataset.iloc[:,1:])

```

```python
for i in train_dataset.columns[1:]:
    train_dataset[i]=train_dataset[i]/np.max(train_dataset[i])
train_dataset

    
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics,model_selection
x=train_dataset.drop("PassengerId",axis=1)

X_train, X_test, y_train, y_test = model_selection.train_test_split(
  x,y , random_state=104,test_size=0.25, shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics,model_selection

clf = DecisionTreeClassifier()


clf = clf.fit(X_train,y_train)


y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
